### ***Exploratory Data Analysis with Python: Air Traffic Data by Farhan***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data from the CSV file
data = pd.read_csv('D:air traffic.csv')
```

```python
data
```

```python
# Remove commas and convert 'RPM' values to float
data['RPM'] = data['RPM'].str.replace(',', '').astype(float)
```

```python
# Now calculate min, max, and mean RPM
min_rpm = data['RPM'].min()
max_rpm = data['RPM'].max()
mean_rpm = data['RPM'].mean()
```

```python
print(f"Minimum RPM: {min_rpm}")
print(f"Maximum RPM: {max_rpm}")
print(f"Mean RPM: {mean_rpm}")
```

```python
# Filter the data for the year range 2008 to 2018
filtered_data = data.query('Year >= 2008 and Year <= 2018')
```

```python
# Display the filtered data
print("\nFiltered Data for 2008 to 2018:")
print(filtered_data)
```

```python
# Remove commas and convert 'ASM' values to float
filtered_data['ASM'] = filtered_data['ASM'].str.replace(',', '').astype(float)
```

```python
# Find the mean ASM of the filtered data
mean_asm_filtered = filtered_data['ASM'].mean()
print(f"\nMean ASM of filtered data: {mean_asm_filtered}")
```

```python
# Select samples for October Month from filtered dataset
october_samples = filtered_data[filtered_data['Month'] == 'October']
```

```python
# Display the samples for October Month
print("\nSamples for October Month:")
print(october_samples)
```

```python
# Remove commas and convert numeric columns to float
numeric_columns = ['Dom_Pax', 'Int_Pax', 'Pax', 'Dom_Flt', 'Int_Flt', 'Flt', 'Dom_RPM', 'Int_RPM', 'RPM', 'Dom_ASM', 'Int_ASM', 'ASM', 'Dom_LF', 'Int_LF', 'LF']
data[numeric_columns] = data[numeric_columns].replace({',': ''}, regex=True).astype(float)
```

```python
# Time Series Line Chart - Example with Dom_Pax
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Dom_Pax', data=data)
plt.title('Domestic Passengers Over Time')
plt.xlabel('Year')
plt.ylabel('Domestic Passengers')
plt.show()
```

```python
# Bar Chart for Monthly Data - Example with Pax
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Pax', data=data)
plt.title('Monthly Total Passengers')
plt.xlabel('Month')
plt.ylabel('Total Passengers')
plt.show()
```

```python
# Box Plot - Example with RPM
plt.figure(figsize=(10, 6))
sns.boxplot(x='Year', y='RPM', data=data)
plt.title('Distribution of RPM Across Years')
plt.xlabel('Year')
plt.ylabel('Revenue Passenger Miles (RPM)')
plt.show()
```

```python
# Correlation Matrix Heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Pie Chart for Passenger Distribution
passenger_distribution = data[['Dom_Pax', 'Int_Pax']].sum()
plt.figure(figsize=(8, 8))
plt.pie(passenger_distribution, labels=passenger_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Passenger Distribution (Domestic vs International)')
plt.show()
```

```python
# Scatter Plot - Example with Dom_RPM vs Dom_ASM
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Dom_RPM', y='Dom_ASM', data=data)
plt.title('Scatter Plot of Domestic RPM vs ASM')
plt.xlabel('Domestic RPM')
plt.ylabel('Domestic ASM')
plt.show()
```

```python
# Stacked Area Chart - Example with Flt
plt.figure(figsize=(10, 6))
plt.stackplot(data['Year'], data['Dom_Flt'], data['Int_Flt'], labels=['Domestic Flights', 'International Flights'], alpha=0.7)
plt.title('Stacked Area Chart of Flights Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Flights')
plt.legend(loc='upper left')
plt.show()
```

```python
# Line Chart for Load Factor (LF) - Example with Dom_LF
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Dom_LF', data=data, label='Domestic LF')
sns.lineplot(x='Year', y='Int_LF', data=data, label='International LF')
plt.title('Load Factor Over Time')
plt.xlabel('Year')
plt.ylabel('Load Factor')
plt.legend()
plt.show()
```
